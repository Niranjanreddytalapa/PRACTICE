```python
df1=sns.load_dataset("tips")
df1.head()
```

```python
sns.boxplot(data=df1)
```

```python
sns.boxplot(x="tip",y="smoker",data=df1)
```

```python
sns.boxplot(x ="tip",y ="smoker",data=df1,hue="sex")
```

```python
sns.set_theme("notebook")
sns.boxplot(x="tip",y="smoker",hue="sex",data=df1)
```

```python
sns.set_theme("notebook")
sns.boxplot(x="tip",y="smoker",hue="sex",data=df1,palette="Set2",linewidth=5)
```

```python
sns.set_theme("talk")
sns.boxplot(x="total_bill",y="sex",data=df1,palette="Set2",linewidth=5)
```
